```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
p1 = pd.read_csv('out/QAgent_p12020_11_15-22_52_18.csv', header=None, names=['optimal'])
```

```python
p1 = p1['optimal'].values
```

```python
#p1 = p1[np.linspace(0,20000,1001, dtype=int)]
N = 100

moving_avg = np.convolve(p1, np.ones((N,))/N, mode='valid')
```

```python
plt.plot(np.arange(len(moving_avg)), moving_avg)
plt.title('Proportion of Optimal Plays When P1')
plt.show()
```

```python
optmoves = pd.read_csv('final/optimal_moves0.10.1-0.1vSelfAll_2020_11_25-16_49_29.csv', header=None, names=np.arange(50000))
```

```python
optmoves = optmoves.values
optmovesmeans = np.mean(optmoves, axis=0)
optmovesmeans = optmovesmeans[~np.isnan(optmovesmeans)]
N = 100
optmovesmeans_moving_avg = np.convolve(optmovesmeans, np.ones((N,))/N, mode='valid')

plt.plot(np.arange(len(optmovesmeans_moving_avg)), optmovesmeans_moving_avg)
plt.show()
```
